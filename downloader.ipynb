```python
from selenium import webdriver
import time

driver  = webdriver.Chrome()
all_downloads = dict()
for i in range(1,40 + 1):
    url = f"https://www.repository.cam.ac.uk/items/6d77cd6d-8569-4bf4-9d5f-311ad2a49ac8/full?obo.page={i}#p-obo"
    driver.get(url)
    driver.implicitly_wait(3)
    print(f"parsing page {i}")
    name_xpaths = ["""//*[@id="p-obo"]/div[2]/div[2]/dl/dd[1]""",
                   """//*[@id="p-obo"]/div[3]/div[2]/dl/dd[1]""",
                   """//*[@id="p-obo"]/div[4]/div[2]/dl/dd[1]""",
                   """//*[@id="p-obo"]/div[5]/div[2]/dl/dd[1]""",
                   """//*[@id="p-obo"]/div[6]/div[2]/dl/dd[1]""",
                  ]
    link_xpaths = ["""//*[@id="p-obo"]/div[2]/div[3]/ds-themed-file-download-link/ds-file-download-link/a""",
                   """//*[@id="p-obo"]/div[3]/div[3]/ds-themed-file-download-link/ds-file-download-link/a""",
                   """//*[@id="p-obo"]/div[4]/div[3]/ds-themed-file-download-link/ds-file-download-link/a""",
                   """//*[@id="p-obo"]/div[5]/div[3]/ds-themed-file-download-link/ds-file-download-link/a""",
                   """//*[@id="p-obo"]/div[6]/div[3]/ds-themed-file-download-link/ds-file-download-link/a""",
                  ]
    for i in range(5):
        try:
            name_elem = driver.find_elements(By.XPATH, name_xpaths[i])
            link_elem = driver.find_elements(By.XPATH, link_xpaths[i])
            
            dl_link = link_elem[0].get_attribute("href")
            dl_code = dl_link.split('/')[4]
            dl_link_actual = f"https://api.repository.cam.ac.uk/server/api/core/bitstreams/{dl_code}/content"
            print(name_elem[0].text, " ", dl_link_actual)
            all_downloads[name_elem[0].text] = dl_link_actual
        except:
            print("failed")
            pass
```

```python
import json
with open('downloads.json', 'w') as fp:
    json.dump(all_downloads,fp)
```

```python
with open("downloads.json", "r") as json_file:
    new_dl_list = json.load(json_file)
```

```python
from subprocess import run, PIPE

for name, link in new_dl_list.items():
    print(f"starting download for {name}")
    command = ["curl", "-o", f"data/{name}", link]
    result = run(command, universal_newlines=True, capture_output=True, shell="True", text=True)
    print(f"finished download for {name}")
    break
```
